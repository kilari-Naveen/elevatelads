```python
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']
data[numeric_cols].hist(figsize=(10, 6), bins=20, edgecolor="black")
plt.suptitle("Histograms of Numeric Features")
plt.show()
```

```python
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']
data[numeric_cols].hist(figsize=(10, 6), bins=20, edgecolor="black")
plt.suptitle("Histograms of Numeric Features")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=data[col], color="skyblue")
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
sns.pairplot(data[['Survived', 'Age', 'Fare', 'SibSp', 'Parch']], hue="Survived")
plt.suptitle("Pairplot of Key Features", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="Pclass", y="Survived", data=data, ci=None)
plt.title("Survival Rate by Passenger Class")
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(x="Sex_male", y="Survived", data=data, ci=None)
plt.title("Survival Rate by Gender (0=Female, 1=Male)")
plt.show()
```
